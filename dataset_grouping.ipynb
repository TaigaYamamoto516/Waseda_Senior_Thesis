```python
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance  import squareform
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def corr_matrix(X, method: str = 'average'):
    # calculate distance
    corr = X.corr().abs().values
    dist = 1 - corr
    np.fill_diagonal(dist, 0)

    dist_vec = squareform(dist)

    # linkage
    linked = linkage(dist_vec, method=method)
    print(f"linkage method: {method}")

    # dendrogram
    plt.figure(figsize=(10, 7))
    dendrogram(linked,
               orientation='top',
               labels=X.columns,
               distance_sort='descending',
               show_leaf_counts=True)
    plt.title('Hierarchical Clustering Dendrogram (1 - |corr| distance)')
    plt.xlabel('Column')
    plt.ylabel('Distance')
    plt.show()

    sns.heatmap(X.corr(), cmap='coolwarm', square=True)
    plt.title('Correlation Matrix of Columns')
    plt.show()

    return linked
```
